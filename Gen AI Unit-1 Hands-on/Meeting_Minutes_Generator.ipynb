## LOAD SUMMARIZATION MODEL

```python
from transformers import pipeline, AutoModelForSeq2SeqLM, AutoTokenizer

summarizer = pipeline(
    "text2text-generation",
    model="sshleifer/distilbart-cnn-12-6"
)
```

## Input Meeting Transcript

```python
meeting_transcript = """
Team Lead: Good morning everyone. Let's start with the Q1 planning session. First, we need to discuss the new feature roadmap.
Developer 1: I think we should prioritize the API redesign. It will improve performance by 40%.
Developer 2: Agreed. We should start by creating technical specifications and allocate it to the backend team.
Manager: That sounds good. Can you have a draft ready by next Friday? Also, the marketing team needs us to launch the mobile app by end of March.
Developer 1: That's tight but doable. We'll need to hire two more developers though.
Manager: I'll work on that. Let's also discuss the database migration project. We need to move from MongoDB to PostgreSQL.
DB Admin: We should schedule a separate meeting for that. But I can say we need at least 2 weeks for planning and testing.
Team Lead: Good. Everyone, please update your task status in Jira by EOD. Our next meeting is on Friday.
Manager: Also, don't forget the team offsite on the 15th. Attendance is mandatory.
"""
```

```python
summary = summarizer(
    meeting_transcript,
    max_length=75,
    min_length=15,
    do_sample=False
)

print("Meeting Minutes:\n")
print(summary[0]["generated_text"])
```

### Observation

The model successfully extracted the main points from the meeting transcript.
Key decisions like API redesign priority, mobile app launch deadline, and database migration planning were captured.
This shows how summarization models can be useful for creating quick action items and meeting notes.
